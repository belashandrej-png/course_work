```python
import matplotlib.pyplot as plt
```

```python
# Импорт библиотеки requests для выполнения HTTP-запросов к API
import requests

# Формирование URL-запроса для получения данных о денежном потоке Netflix через Alpha Vantage API
url = 'https://www.alphavantage.co/query?function=CASH_FLOW&symbol=NFLX&apikey=свойAPI'

# Отправка GET-запроса к API и сохранение ответа в переменную inf
inf = requests.get(url)
```

```python
# Проверка типа объекта
type(inf)
```

```python
# Преобразование ответа в формат JSON
inf.json()
```

```python
# Сохранение JSON-данных в переменную data
data = inf.json()
data
```

```python
# Извлечение годовых отчетов из полученных данных
# По ключу 'annualReports' из словаря data извлекается список годовых финансовых отчетов
annual_reports = data['annualReports']
annual_reports
```

```python
import pandas as pd

df = pd.DataFrame(annual_reports)
df
```

```python
columns_nes = ['fiscalDateEnding', 'operatingCashflow', 'capitalExpenditures', 'dividendPayout']
df_choise = df[columns_nes]
df_choise
```

```python
df_choise.columns = ['Дата отчетности', 'Операционный денежный поток', 'Капитальные затраты', 'Дивидендные выплаты']
df_choise
```

```python
df_2 = df_choise.copy()
```

```python
df_2['Дата отчетности'] = pd.to_datetime(df_2['Дата отчетности'])

if df_2['Дивидендные выплаты'].dtype == 'object':

    df_2['Дивидендные выплаты'] = df_2['Дивидендные выплаты'].replace('None', '0')

    df_2['Дивидендные выплаты'] = df_2['Дивидендные выплаты'].astype(float)


df_2 = df_2.sort_values('Дата отчетности')


print("\nТипы данных:")
print(df_2.dtypes)
df_2
```

```python
df_2['Скользящее среднее (2 года)'] = df_2['Операционный денежный поток'].rolling(window=2).mean()
df_2['Скользящее среднее (3 года)'] = df_2['Операционный денежный поток'].rolling(window=3).mean()

df_2
```

```python
plt.figure(figsize=(15, 5))
bars = plt.bar(df_2['Дата отчетности'], df_2['Дивидендные выплаты'],
              color='green', alpha=0.7, edgecolor='black')

for bar in bars:
    height = bar.get_height()
    if height != 0:
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'${height:,.0f}',
                ha='center', va='bottom', fontsize=9, rotation=45)

plt.title('Дивидендные выплаты Netflix', fontsize=16, fontweight='bold')
plt.xlabel('Год', fontsize=12)
plt.ylabel('Значение ($)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(['Дивидендные выплаты'], loc='best', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Расчет свободного денежного потока
numeric_columns = ['Операционный денежный поток', 'Капитальные затраты']
for col in numeric_columns:
    if df_2[col].dtype == 'object':
        df_2[col] = df_2[col].str.replace('$', '').str.replace(',', '').astype(float)

# Свободный денежный поток
df_2['Свободный денежный поток'] = df_2['Операционный денежный поток'] + df_2['Капитальные затраты']

plt.figure(figsize=(15, 7))
plt.bar(df_2['Дата отчетности'].dt.year.astype(str), df_2['Свободный денежный поток'],
        color='purple', alpha=0.8)

for i, v in enumerate(df_2['Свободный денежный поток']):
    if abs(v) >= 1e9:
        label = f'${v/1e9:,.1f}B'
    elif abs(v) >= 1e6:
        label = f'${v/1e6:,.1f}M'
    else:
        label = f'${v:,.0f}'

    plt.text(i, v + (0.1 * max(abs(df_2['Свободный денежный поток']))),
             label,
             ha='center', va='bottom', fontsize=10)

plt.title('Свободный денежный поток Netflix', fontsize=16, fontweight='bold')
plt.xlabel('Год', fontsize=12)
plt.ylabel('Доллары ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
